```python
from groq import Groq
from dotenv import load_dotenv
import os

# Load the API key from .env file
load_dotenv()

# Create the Groq client using the API key
client = Groq(api_key=os.getenv("GROQ_API_KEY"))

# The AI model we want to use
model = "llama-3.3-70b-versatile"


# ── CONVERSATION HISTORY ──────────────────────────────────────────
# We keep a list called 'messages' to store the full conversation.
# Each message is a dictionary with two keys:
#   "role"    → who sent it: "user" (you) or "assistant" (AI)
#   "content" → the actual text of the message
#
# Example of what messages looks like after 1 exchange:
# [
#   {"role": "user",      "content": "Hello!"},
#   {"role": "assistant", "content": "Hi! How can I help?"}
# ]
# We send this ENTIRE list every time so the AI remembers the chat.
# ──────────────────────────────────────────────────────────────────


def add_user_message(messages, text):
    """Add YOUR message to the conversation history."""
    user_message = {
        "role": "user",    # identifies this as your message
        "content": text    # the text you typed
    }
    messages.append(user_message)


def add_assistant_message(messages, text):
    """Add the AI's reply to the conversation history."""
    assistant_message = {
        "role": "assistant",  # identifies this as AI's message
        "content": text       # the text the AI replied
    }
    messages.append(assistant_message)


def chat(messages):
    """Send the full conversation to Groq and get the AI's reply."""
    
    response = client.chat.completions.create(
        model=model,          # which AI model to use
        max_tokens=1000,      # maximum length of the reply
        messages=messages,    # the full conversation history
    )
    
    # Extract just the text from the response object
    # response.choices[0] → first (and only) reply
    # .message.content    → the actual text
    return response.choices[0].message.content


# ── HOW TO USE ────────────────────────────────────────────────────

# Step 1: Start with an empty conversation
messages = []

# Step 2: Add your message
add_user_message(messages, "My name is Priya. What is quantum computing? Answer in one sentence")

# Step 3: Get AI's reply
reply = chat(messages)
print("AI:", reply)

# Step 4: Save AI's reply to history (so it remembers)
add_assistant_message(messages, reply)

# Step 5: Ask a follow-up — AI will remember context from above
add_user_message(messages, "Can you summarize that in one sentence?")
reply = chat(messages)
print("AI:", reply)
add_assistant_message(messages, reply)

# Step 6: Test memory
add_user_message(messages, "What is my name?")
reply = chat(messages)
print("AI:", reply)  # Should say "Priya" because it remembers!
```
